# TMDB - WEB SCRAPPING

```python
import requests
import pandas as pd

api_key = 'Your key'
```

```python
#Test scrapping
url = f'https://api.themoviedb.org/3/discover/movie?api_key={api_key}&with_original_language=ta'

response = requests.get(url)
data = response.json()

movies = []

for movie in data['results']:
    movie_details = {
        'Title': movie['title'],
        'Rating': movie['vote_average'],
        'Release Year': movie['release_date'][:4],
        'Overview': movie['overview']
    }
    movies.append(movie_details)

df = pd.DataFrame(movies)
print(df)
```

Tamil movies

```python
# Fetch genre information
genre_url = f'https://api.themoviedb.org/3/genre/movie/list?api_key={api_key}&language=en-US'
genre_response = requests.get(genre_url)
genres = genre_response.json()['genres']
genre_dict = {genre['id']: genre['name'] for genre in genres}

# Fetch Tamil movies
movies = []
page = 1

while True:
    url = f'https://api.themoviedb.org/3/discover/movie?api_key={api_key}&with_original_language=ta&page={page}'
    response = requests.get(url)
    data = response.json()

    if 'results' not in data or not data['results']:
        break

    for movie in data['results']:
        genre_names = [genre_dict[genre_id] for genre_id in movie['genre_ids']]
        movie_details = {
            'Title': movie['title'],
            'Rating': movie['vote_average'],
            'Release Year': movie['release_date'][:4],
            'Genres': ', '.join(genre_names)
        }
        movies.append(movie_details)

    page += 1

df = pd.DataFrame(movies)
print(df)
```

```python
df.to_csv('tamil_movies.csv', index=False)
```
