```python
import pandas  as pd
import numpy as np
```

```python
df_hi = pd.read_csv('/content/hindi_movies.csv')
df_ta = pd.read_csv('/content/tamil_movies.csv')
df_te = pd.read_csv('/content/telugu_movies.csv')
```

```python
df_all=pd.concat([df_hi, df_ta, df_te], ignore_index=True)
```

```python
df_all.shape
```

```python
df_all.info()
```

```python
df_all.head()
```

```python
df_all = df_all.drop(columns=['Unnamed: 5'])
```

```python
df_all.head()
```

```python
df_all['Release Year'].unique().sort()
```

```python
df_all['Rating'].unique()
```

```python
movies_zero_rating = df_all[df_all['Rating'] == 0.0]
print(movies_zero_rating)
```

```python
df_all= df_all[df_all['Rating'] != 0.0]
print(df_all)
```

```python
df_all.isnull().sum()
```

```python
df_all = df_all[df_all['Release Year'] != 2024.0]
```

```python
df_all = df_all.dropna()
```

```python
df_all
```

```python
df_all = df_all.drop_duplicates(subset='Title')
```

```python
df_all
```

```python
df_all['Release Year'] = df_all['Release Year'].astype(int)
```

```python
movies_per_year = df_all['Release Year'].value_counts().sort_index()
```

```python
print(movies_per_year)
```

```python
# Drop rows where Release Year is before 1970
df_all = df_all[df_all['Release Year'] >= 1970]
```

```python
df_all.to_csv('preprocessed.csv', index=False)
```
